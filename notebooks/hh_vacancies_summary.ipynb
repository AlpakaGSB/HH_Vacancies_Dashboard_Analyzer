## Ключевые выводы (на основе публичных данных)

- **Объём рынка**
  IT: 8 256 вакансий
  Курьеры: 3 824 вакансий
  → IT-рынок в ~2.2 раза активнее

- **Удалённая работа**
  IT — 43.9 %
  Курьеры — 0.4 %
  → Почти половина IT можно делать из дома, курьерская работа — 99.6 % офлайн, что, конечно, логично

- **Зарплата указана**
  IT — 35.1 % вакансий
  Курьеры — 98.7 % вакансий
  → У курьеров ЗП почти всегда прозрачна, в IT чаще «по договорённости»

- **Опыт работы**
  Курьеры: 92.4 % — без опыта
  IT: 85 % — 1–6 лет (мидлы доминируют), всего 5.7 % без опыта

- **Топ-компании**
  Ozon — лидер в обеих категориях (особенно у курьеров — 807 вакансий)

- **Частые слова в описании**
  IT: опыт, данных, работы, python, знание, разработка, аналитики, sql
  Курьеры: заказы, доставлять, доставка, клиентов, умение, ип, пунктуальность, самокат, велосипед
